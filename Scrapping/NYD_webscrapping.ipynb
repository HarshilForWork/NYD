```python
import requests
from bs4 import BeautifulSoup
import time
import os
import re

def scrape_sarga(sarga_number):
    """Scrape content from a specific sarga in Kishkinda Kanda with proper pratipada extraction"""
    url = f"https://www.valmikiramayan.net/utf8/kish/sarga{sarga_number}/kishkindhasans{sarga_number}.htm"

    try:
        print(f"Fetching URL: {url}")
        response = requests.get(url)
        response.raise_for_status()

        soup = BeautifulSoup(response.text, 'html.parser')
        results = []

        # Find all paragraphs
        all_paragraphs = soup.find_all('p')

        # Track current verse number
        current_shloka_num = None
        current_verse_text = None

        # Devanagari detection pattern
        devanagari_pattern = re.compile(r'[\u0900-\u097F]')

        # Special pattern to extract shloka numbers from pratipada sections
        # Looking for patterns like "73. Soumitri;" or "74. dhaatubhiH ="
        pratipada_pattern = re.compile(r'^\s*(\d+)\.\s+(.+?)\s*=\s*')

        for p in all_paragraphs:
            text = p.text.strip()
            if not text:
                continue

            classes = p.get('class', [])

            # Check if this is a Sanskrit verse (contains Devanagari or has 'tat' class)
            if 'tat' in classes or devanagari_pattern.search(text):
                # Store as a Sanskrit verse
                current_verse_text = text
                # Reset the current shloka number as we'll find it in the pratipada that follows
                current_shloka_num = None

                # Store the verse with a temporary number (will be updated if we find pratipada)
                results.append({
                    'type': 'shloka',
                    'temp_id': len(results),  # Use position as temporary ID
                    'number': None,  # Will be filled in later
                    'text': text
                })

            # Check if this is a pratipada (word-by-word translation)
            # Typically follows right after the Sanskrit verse
            elif not current_shloka_num and ('pra' in classes or pratipada_pattern.search(text)):
                # Try to extract shloka number from pratipada
                match = pratipada_pattern.search(text)
                if match:
                    current_shloka_num = match.group(1)
                    print(f"Found shloka number {current_shloka_num} from pratipada")

                    # Update the previous verse with this number
                    for i in range(len(results)-1, -1, -1):
                        if results[i]['type'] == 'shloka' and results[i]['number'] is None:
                            results[i]['number'] = current_shloka_num
                            break

            # Check if this is a commentary
            elif 'comment' in classes or text.startswith('[Commentary]'):
                results.append({
                    'type': 'comment',
                    'number': current_shloka_num,
                    'text': text
                })

            # Check if this is a translation
            elif 'txt' in classes or (len(text) > 30 and '"' in text):
                results.append({
                    'type': 'translation',
                    'number': current_shloka_num,
                    'text': text
                })

            # Check if this contains "Verse Locator" - marks end of a section
            if "Verse Locator" in text:
                current_shloka_num = None

        # Second pass to fill in any missing numbers using sequence
        shloka_counter = 1
        for item in results:
            if item['type'] == 'shloka' and item['number'] is None:
                item['number'] = str(shloka_counter)
                shloka_counter += 1

        # Remove temporary IDs
        for item in results:
            if 'temp_id' in item:
                del item['temp_id']

        return results

    except requests.exceptions.RequestException as e:
        print(f"Error scraping sarga {sarga_number}: {e}")
        return []

def extract_shloka_number_from_text(text):
    """Try to extract shloka number from various text formats"""
    # Try pattern like "73. Soumitri; pashya ="
    match = re.search(r'^\s*(\d+)\.\s+', text)
    if match:
        return match.group(1)

    # Try pattern like "Verse 4-1-73"
    match = re.search(r'Verse\s+\d+-\d+-(\d+)', text)
    if match:
        return match.group(1)

    return None

def main():
    # Create a debug log file
    with open("scraper_debug.log", "w", encoding="utf-8") as debug_file:
        debug_file.write("Scraper Debug Log\n")
        debug_file.write("=" * 50 + "\n\n")

    # Create or open the file to write
    with open("Kishkinda_Kanda.txt", "w", encoding="utf-8") as file:
        file.write("KISHKINDA KANDA\n")
        file.write("=" * 50 + "\n\n")

        # For initial testing, focus on sarga 1 where we know the issue exists
        # total_sargas = 67
        # Adjust this range to test specific sargas
        sarga_range = [1]  # Start with just sarga 1

        # Iterate through sargas
        for sarga_num in sarga_range:
            try:
                print(f"Processing Sarga {sarga_num}...")
                file.write(f"\n\nSARGA {sarga_num}\n")
                file.write("-" * 30 + "\n\n")

                # Scrape the content
                content = scrape_sarga(sarga_num)

                # Log debug info
                with open("scraper_debug.log", "a", encoding="utf-8") as debug_file:
                    debug_file.write(f"\n\nSARGA {sarga_num}\n")
                    debug_file.write("-" * 30 + "\n\n")
                    debug_file.write(f"Total items found: {len(content)}\n")

                    # Log each item for debugging
                    for i, item in enumerate(content):
                        debug_file.write(f"Item {i+1}: {item['type']}, Number: {item.get('number')}\n")
                        debug_file.write(f"Text preview: {item['text'][:50]}...\n\n")

                # Track which shloka numbers we've already seen
                seen_shloka_numbers = set()

                if content:
                    # Write content to file
                    for item in content:
                        if item["type"] == "shloka":
                            shloka_num = item.get("number", "?")

                            # Avoid duplicate shloka numbers
                            if shloka_num in seen_shloka_numbers:
                                continue
                            seen_shloka_numbers.add(shloka_num)

                            file.write(f"Shlok {shloka_num}: {item['text']}\n\n")

                        elif item["type"] == "comment":
                            file.write(f"[Commentary] {item['text']}\n\n")

                        elif item["type"] == "translation":
                            file.write(f"[Translation] {item['text']}\n\n")
                else:
                    file.write(f"[No content found for Sarga {sarga_num}]\n")

                # Add a small delay
                time.sleep(1)

            except Exception as e:
                print(f"Error processing sarga {sarga_num}: {e}")
                with open("scraper_debug.log", "a", encoding="utf-8") as debug_file:
                    debug_file.write(f"ERROR PROCESSING SARGA {sarga_num}: {str(e)}\n")
                file.write(f"[Error processing Sarga {sarga_num}]\n")

    print(f"Completed! Content saved to Kishkinda_Kanda.txt")
    print(f"Debug log saved to scraper_debug.log")
    print(f"Current directory: {os.getcwd()}")

# Function to help inspect a specific sarga in detail
def inspect_sarga(sarga_number):
    """Download and analyze the HTML structure of a specific sarga"""
    url = f"https://www.valmikiramayan.net/utf8/kish/sarga{sarga_number}/kishkindhasans{sarga_number}.htm"

    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        # Save the raw HTML for inspection
        with open(f"sarga_{sarga_number}_raw.html", "w", encoding="utf-8") as f:
            f.write(soup.prettify())

        # Extract all paragraphs and their classes
        all_paragraphs = soup.find_all('p')

        with open(f"sarga_{sarga_number}_structure.txt", "w", encoding="utf-8") as f:
            f.write(f"Structure Analysis for Sarga {sarga_number}\n")
            f.write("=" * 50 + "\n\n")

            for i, p in enumerate(all_paragraphs):
                classes = p.get('class', [])
                text = p.text.strip()
                if text:
                    f.write(f"Paragraph {i+1}:\n")
                    f.write(f"Classes: {classes}\n")
                    f.write(f"Text preview: {text[:100]}...\n\n")

        print(f"Inspection of Sarga {sarga_number} complete. See sarga_{sarga_number}_structure.txt for details.")

    except Exception as e:
        print(f"Error inspecting sarga {sarga_number}: {e}")

if __name__ == "__main__":
    # Uncomment to run inspection on a specific sarga
    # inspect_sarga(1)

    # Run the main scraper
    main()
```

```python
#NEW FINAL CODE (HOPEFULLY)
import requests
from bs4 import BeautifulSoup
import time
import os

def scrape_sarga(sarga_number):
    """Scrape content from a specific sarga number in Sundara Kanda"""
    url=f"https://www.valmikiramayan.net/utf8/yuddha/sarga{sarga_number}/yuddhasans{sarga_number}.htm"

    try:
        response = requests.get(url)
        response.raise_for_status()  # Check if request was successful

        # Create a Beautiful Soup object
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all paragraphs with class "tat" or "comments"
        all_content = []

        # Find all paragraphs regardless of class
        all_paragraphs = soup.find_all("p")

        # Filter and keep only those with relevant classes
        for p in all_paragraphs:
            classes = p.get("class", [])
            if "tat" in classes or "comment" in classes or "pratipada" in classes or "txt" in classes:
                # Store the text content and its class for reference
                if "tat" in classes:
                    content_type = "tat"
                elif "comment" in classes:
                    content_type = "comment"
                elif "pratipada" in classes:
                    content_type = "pratipada"
                else:
                    content_type = "txt"
                all_content.append({"type": content_type, "text": p.text.strip()})

        return all_content

    except requests.exceptions.RequestException as e:
        print(f"Error scraping sarga {sarga_number}: {e}")
        return []

def main():
    # Create or open the file to write
    with open("Yuddha_Kanda.txt", "w", encoding="utf-8") as file:
        file.write("Yuddha Kanda\n")
        file.write("=" * 50 + "\n\n")

        # The Ayodhya Kanda has 119 sargas
        total_sargas = 128

        # Iterate through all sargas (1-119)
        for sarga_num in range(1, total_sargas + 1):
            try:
                print(f"Processing Sarga {sarga_num}...")
                file.write(f"\n\nSARGA {sarga_num}\n")
                file.write("-" * 30 + "\n\n")

                # Scrape the content
                content = scrape_sarga(sarga_num)

                if content:
                    # Counter for "tat" paragraphs within this sarga
                    shlok_counter = 1
                    current_shlok_number = None

                    # Write content to file
                    for item in content:
                        if item["type"] == "comment":
                            file.write("[Commentary] " + item["text"] + "\n\n")
                        elif item["type"] == "txt":
                            file.write("[Text] " + item["text"] + "\n\n")
                        elif item["type"] == "pratipada":
                            # Try to extract shloka numbers from the text
                            text = item["text"]
                            # Look for patterns like "1-2", "42,43", "41,42a", "41b,42c"
                            import re
                            numbers = re.match(r'^\s*(\d+[a-z]?(?:[-,;]\s*\d+[a-z]?)*)', text)

                            if numbers:
                                current_shlok_number = numbers.group(1)
                                # Replace comma with hyphen for consistent notation
                                current_shlok_number = current_shlok_number.replace(',', '-').replace(';', '-')
                                # We don't write the pratipada text, just store the number for next tat paragraph
                            # If no numbers found or if we don't want pratipada text, we skip it entirely
                        else:  # This is a "tat" class paragraph
                            if current_shlok_number:
                                # Use the extracted number from the previous pratipada
                                file.write(f"Shlok {current_shlok_number}: {item['text']}\n\n")
                                current_shlok_number = None
                            else:
                                # Print if no number
                                file.write("\n\n"+item["text"]+"\n\n")

                else:
                    file.write(f"[No content found for Sarga {sarga_num}]\n\n")

                # Add a small delay to avoid overwhelming the server


            except Exception as e:
                print(f"Error processing sarga {sarga_num}: {e}")
                file.write(f"[Error processing Sarga {sarga_num}]\n")

    print(f"Completed! Content saved to Yuddha_Kanda.txt")
    print(f"Current directory: {os.getcwd()}")

if __name__ == "__main__":
    main()
```
